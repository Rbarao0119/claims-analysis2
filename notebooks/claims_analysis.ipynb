```python
# 1. Load Libraries
# =====================================
import pandas as pd
import numpy as np
```

```python
# =====================================
# 2. Upload or Load CSV Files
from google.colab import files

print("Upload HEADER file")
uploaded = files.upload()
df_header = pd.read_csv(list(uploaded.keys())[0])

print("Upload LINE file")
uploaded = files.upload()
df_line = pd.read_csv(list(uploaded.keys())[0])

print("Upload CODE file")
uploaded = files.upload()
df_code = pd.read_csv(list(uploaded.keys())[0])
```

```python
# 3. Explore Each File
# =====================================
def explore(df, name):
    print("="*80)
    print(f"ðŸ“Œ Exploring: {name}")
    print("="*80)
    print("Shape:", df.shape)
    print("\nFirst 5 rows:")
    display(df.head())
    print("\nColumn info:")
    display(df.dtypes)
    print("\nMissing values:")
    display(df.isna().sum())
    print("\nBasic numeric statistics:")
    display(df.describe())

explore(df_header, "HEADER")
explore(df_line, "LINE")
explore(df_code, "CODE")
```

```python
# 4. Key Observations & Calculations

# ---- A. Number of unique claims ----
unique_claims = df_header['ProspectiveClaimId'].nunique()
print("\n Unique Claims:", unique_claims)
```

```python
# ---- B. Date Range of Claims ----
# (Assuming column names: 'claim_date' or similar â€” update if needed)
date_col = [col for col in df_header.columns if "date" in col.lower()][0]

df_header[date_col] = pd.to_datetime(df_header[date_col], errors='coerce')

date_min = df_header[date_col].min()
date_max = df_header[date_col].max()

print("\n Claim Date Range:")
print("Earliest:", date_min)
print("Latest:", date_max)
```

```python
# ---- C. Average service lines per claim ----
avg_lines = df_line.groupby("ProspectiveClaimId").size().mean()
print("\n Average service lines per claim:", round(avg_lines, 2))
```

```python
# ---- D. Average diagnosis codes per claim ----
avg_dx = df_code.groupby("ProspectiveClaimId").size().mean()
print("\n Average diagnosis codes per claim:", round(avg_dx, 2))
```

```python
import matplotlib.pyplot as plt

# Ensure plots display nicely
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# QUESTION 1

# Create a 'BillingProviderFullName' column for display purposes, handling NaNs
df_header['BillingProviderFullName'] = df_header['BillingProvFirstName'].fillna('') + ' ' + df_header['BillingProvLastName'].fillna('')
df_header['BillingProviderFullName'] = df_header['BillingProviderFullName'].str.strip()

provider_counts = (
    df_header.groupby(['BillingProviderNPI', 'BillingProviderFullName'])['ProspectiveClaimId']
    .nunique()
    .reset_index(name='claim_count')
    .sort_values('claim_count', ascending=False)
)

top5_providers = provider_counts.head(5)
display(top5_providers)

# Bar chart
plt.bar(top5_providers['BillingProviderFullName'], top5_providers['claim_count'])
plt.xticks(rotation=45, ha='right')
plt.title("Top 5 Billing Providers by Number of Claims")
plt.xlabel("Provider Name")
plt.ylabel("Claim Count")
plt.show()
```

```python
# QUESTION 2

# Count claims by payer (using ProspectiveClaimId)
payer_counts = (
    df_header.groupby('PrimaryPayerName')['ProspectiveClaimId']
    .nunique()
    .reset_index(name='claim_count')
    .sort_values('claim_count', ascending=False)
)

# Add percentage of total claims
payer_counts['percent'] = (
    payer_counts['claim_count'] / payer_counts['claim_count'].sum() * 100
)

top5_payers = payer_counts.head(5)
display(top5_payers)

# Bar chart
plt.bar(top5_payers['PrimaryPayerName'], top5_payers['claim_count'])
plt.xticks(rotation=45, ha='right')
plt.title("Top 5 Primary Payers by Claim Volume")
plt.xlabel("Primary Payer")
plt.ylabel("Claim Count")
plt.show()

# Pie chart for full payer mix
plt.pie(
    payer_counts['claim_count'],
    labels=payer_counts['PrimaryPayerName'],
    autopct='%1.1f%%'
)
plt.title("Overall Payer Mix")
plt.show()
```

```python
# QUESTION 3 â€” Most Common Diagnosis Codes

# Count diagnosis codes from df_code
dx_counts = (
    df_code['CodeValue']
    .value_counts()
    .reset_index()
)
dx_counts.columns = ['ICD10_Code', 'frequency']

top10_dx = dx_counts.head(10)
display(top10_dx)
```

```python
# QUESTION 4 â€” Most Common Procedure Codes

# Auto-detect procedure column in df_line
proc_col = None
for col in df_line.columns:
    if col.lower() in ['procedurecode', 'hcpcs', 'cpt']:
        proc_col = col
        break

if proc_col is None:
    raise ValueError("No procedure code column found. Tell me the exact column name.")

procedure_counts = (
    df_line[proc_col]
    .value_counts()
    .reset_index()
)
procedure_counts.columns = ['ProcedureCode', 'frequency']

top10_proc = procedure_counts.head(10)
display(top10_proc)

# Bar chart
plt.bar(top10_proc['ProcedureCode'], top10_proc['frequency'])
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Most Common Procedure Codes")
plt.xlabel("Procedure Code")
plt.ylabel("Frequency")
plt.show()
```

```python
# QUESTION 5 â€” Place of Service Analysis

# Count claims by PlaceOfService
pos_counts = (
    df_header.groupby('PlaceOfService')['ProspectiveClaimId']
    .nunique()
    .reset_index(name='claim_count')
)
pos_counts['percent'] = (
    pos_counts['claim_count'] / pos_counts['claim_count'].sum() * 100
)

display(pos_counts)

# Compute specific categories using numeric PlaceOfService codes
# Common POS codes: 11 (Office), 21 (Inpatient Hospital), 22 (Outpatient Hospital)
inpatient_pct = pos_counts.loc[
    pos_counts['PlaceOfService'] == 21,
    'percent'
].sum()

office_pct = pos_counts.loc[
    pos_counts['PlaceOfService'] == 11,
    'percent'
].sum()

print("\nINPATIENT % of all claims:", round(inpatient_pct, 2))
print("DOCTOR'S OFFICE % of all claims:", round(office_pct, 2))
```

```python
# Question 6

charge_col = None
for col in df_line.columns:
    if 'charge' in col.lower() or 'amount' in col.lower():
        charge_col = col
        break

if charge_col is None:
    raise ValueError("No charge column found. Tell me the correct name (e.g., LineCharge, ChargeAmount).")

# =======================================
# Merge HEADER + LINE on ProspectiveClaimId
# =======================================
merged_hl = df_line.merge(
    df_header[['ProspectiveClaimId', 'BillingProviderNPI', 'BillingProviderFullName']],
    on='ProspectiveClaimId',
    how='left'
)

# Count service lines per claim
line_counts = (
    merged_hl.groupby(['ProspectiveClaimId', 'BillingProviderFullName'])
    .agg(
        num_lines=('ProspectiveClaimId', 'count'),
        total_charges=(charge_col, 'sum')
    )
    .reset_index()
)

# Filter claims with 5+ service lines
high_line_claims = line_counts[line_counts['num_lines'] >= 5]

display(high_line_claims.head(20))
```

```python
# Question 7

# Step 1: Merge LINE with CODE on ProspectiveClaimId
# =======================================
line_code = df_line.merge(
    df_code[['ProspectiveClaimId', 'CodeValue']],
    on='ProspectiveClaimId',
    how='left'
)

# =======================================
# Step 2: Filter for CPT/HCPCS 99291
# =======================================
proc_col = None
for col in df_line.columns:
    if col.lower() in ['procedurecode', 'cpt', 'hcpcs']:
        proc_col = col
        break

if proc_col is None:
    raise ValueError("No procedure code column found in df_line.")

dx_99291 = line_code[line_code[proc_col] == '99291']

# Count diagnosis codes
dx_counts_99291 = (
    dx_99291['CodeValue']
    .value_counts()
    .reset_index()
)
dx_counts_99291.columns = ['DiagnosisCode', 'Count']

display(dx_counts_99291.head(10))
```

```python
# Question 8

hl = df_line.merge(
    df_header[['ProspectiveClaimId', 'PrimaryPayerName']],
    on='ProspectiveClaimId',
    how='left'
)

# Calculate charges per claim
claim_charges = (
    hl.groupby(['PrimaryPayerName', 'ProspectiveClaimId'])
    .agg(
        total_claim_charge=(charge_col, 'sum')
    )
    .reset_index()
)

# Aggregate by payer
payer_charges = (
    claim_charges.groupby('PrimaryPayerName')
    .agg(
        total_charges=('total_claim_charge', 'sum'),
        avg_charge_per_claim=('total_claim_charge', 'mean'),
        num_claims=('ProspectiveClaimId', 'nunique')
    )
    .sort_values('total_charges', ascending=False)
    .reset_index()
)

display(payer_charges.head(10))
```

```python
## Question 9: â€œWhich providers bill for the most complex cases?â€

# Merge df_header with diagnosis codes
header_code = df_header.merge(
    df_code[['ProspectiveClaimId', 'CodeValue']],
    on='ProspectiveClaimId',
    how='left'
)

# Count diagnoses per claim
dx_per_claim = (
    header_code.groupby(['ProspectiveClaimId', 'BillingProviderFullName'])
    .agg(num_dx=('CodeValue', 'count'))
    .reset_index()
)

# Average diagnosis count per provider
provider_complexity = (
    dx_per_claim.groupby('BillingProviderFullName')
    .agg(avg_dx_per_claim=('num_dx', 'mean'),
         total_claims=('ProspectiveClaimId', 'nunique'))
    .sort_values('avg_dx_per_claim', ascending=False)
)

display(provider_complexity.head(10))

# Visualization
top10_complex = provider_complexity.head(10)

plt.bar(top10_complex.index, top10_complex['avg_dx_per_claim'])
plt.xticks(rotation=45, ha='right')
plt.title("Providers With the Most Complex Cases (Avg Diagnoses per Claim)")
plt.ylabel("Avg Diagnosis Codes")
plt.xlabel("Provider")
plt.show()
```
